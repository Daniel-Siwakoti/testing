```python

import matplotlib.pyplot as plt
f=open('F:/new Downloads/comsol/charge vs voltage (1).txt','r')
voltage=[]
charge=[]
for line in f:
       if line.startswith("%") == False:
            series=line.split()               
            voltage.append(series[0])
            charge.append(series[1])
            
voltagee=list(map(float,voltage))
chargee=list(map(float,charge))

count=0
dq=[]
dv=[]
capacitance=[]

for i in voltagee:
    if i==voltagee[0]:
                   
        dq.append(chargee[voltagee.index(i)+1]-chargee[voltagee.index(i)])
        dv.append(voltagee[voltagee.index(i)+1]-voltagee[voltagee.index(i)])
        capacitance.append(dq[voltagee.index(i)]/dv[voltagee.index(i)])     
            
    elif i>voltagee[0] and i<voltagee[len(voltagee)-1]:
        dq.append(chargee[voltagee.index(i)+1]-chargee[voltagee.index(i)-1])
        dv.append(voltagee[voltagee.index(i)+1]-voltagee[voltagee.index(i)-1])
        capacitance.append(dq[voltagee.index(i)]/dv[voltagee.index(i)])

    else:

        dq.append(chargee[voltagee.index(i)]-chargee[voltagee.index(i)-1])
        dv.append(voltagee[voltagee.index(i)]-voltagee[voltagee.index(i)-1])
        capacitance.append(dq[voltagee.index(i)]/dv[voltagee.index(i)])


        # dq.append(chargee[voltagee.index(i)])
        # dv.append(voltagee[voltagee.index(i)])
        # capacitance.append(dq[voltagee.index(i)]/dv[voltagee.index(i)])     
        

plt.plot(voltagee,capacitance,label='capacitance')
plt.plot(voltagee,chargee,label='charge')
plt.xlabel('voltage')
plt.legend()
plt.show()
f.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
f=open('E:/Thesis/reports/pn2 cv/file4_5v.txt','r')
voltage=[]
charge=[]
for line in f:
       if line.startswith("%") == False:
            series=line.split()               
            voltage.append(series[0])
            charge.append(series[1])
voltagee=list(map(float,voltage))
chargee=list(map(float,charge))
capa=[]
dy=np.diff(chargee)
dx=np.diff(voltagee)
dx=list(dx)
dy=list(dy)
for  i in range(len(chargee)):
    if i<len(chargee)-1:
     capa.append(dy[i]/dx[i])
    else:
     capa.append(chargee[i]/voltagee[i]) 
   
plt.plot(voltagee,capa,label='capacitance')
plt.plot(voltagee,chargee,label='charge')
plt.xlabel('voltage')
plt.legend()
plt.show()
f.close()
```
